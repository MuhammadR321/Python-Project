```python
import pandas as pd
import numpy as np

#graphing data
import matplotlib.pyplot as plt
plt.style.use('classic')

data_source_file = 'BrentOilPrices.csv'
df_pop = pd.read_csv(data_source_file)
df_pop.info()

df_pop.head(100)
```

```python
df_pop_by_date = df_pop.set_index('Date')
df_pop_by_date.plot(xlabel = 'Date', ylabel = 'Price')
plt.yticks(rotation = 45)
plt.xticks(rotation = 90)
```
